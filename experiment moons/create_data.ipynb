```python
from sklearn.datasets import make_moons
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X, _ = make_moons(n_samples=3000, noise=0.05, random_state=42)
```

```python
df = pd.DataFrame(X, columns=['x1', 'x2'])
df.head()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X[:, 0], X[:, 1], s=10, edgecolor='k', alpha=0.7)
plt.title('  make_moons')
plt.xlabel('x1')
plt.ylabel('x2')
plt.grid(True)
plt.show()
```

```python
df.to_csv('data/moons_dataset.csv', index=False)
```
