```python
import pandas as pd
from sklearn.datasets import make_moons
```

```python
def process_data(df, categorical_cols, numeric_cols):

    # One-Hot Encoding
    df_processed = pd.get_dummies(df, columns=categorical_cols, drop_first=False)

    for col in df_processed.columns:
        if df_processed[col].dtype == 'bool':
            df_processed[col] = df_processed[col].astype(int)

    #CDF
    for col in numeric_cols:
        df_processed[col] = df_processed[col].rank() / len(df_processed)

    return df_processed
```

**Insurance**<br>
Num: 4 Cat: 3

```python
df_insurance = pd.read_csv('data/original/insurance.csv')

len(df_insurance)
```

```python
df_insurance.isna().sum(axis=0)
```

```python
df_insurance.head()
```

```python
categorical_cols = ['sex', 'smoker', 'region']
numeric_cols = ['age', 'bmi', 'children', 'expenses']

df_processed = process_data(df_insurance, categorical_cols, numeric_cols)
df_processed.head()
```

```python
df_processed.shape
```

```python
df_processed.to_csv('data/processed/insurance.csv', index=False)
```
